```
# IMPORTING THE DATASET"""

dataset = pd.read_csv("/content/apple_quality.csv")
```

```
# Handling missing values
# Checking for missing values
print(dataset.isnull().sum())
```

```
# Checking unique values in 'Acidity' column
print(dataset['Acidity'].unique())
```

```
# Removing rows with non-numeric values in 'Acidity' column
dataset = dataset[dataset['Acidity'].apply(lambda x: str(x).replace('.', '', 1).lstrip('-').isdigit())]
```

```
# Now, we can calculate the mean and replace missing values
mean_acidity = dataset['Acidity'].astype(float).mean()
dataset['Acidity'].fillna(mean_acidity, inplace=True)
```

```
# Confirming missing values have been handled
print(dataset.isnull().sum())
```

```
# Converting categorical variables into numerical format
# For 'Quality' column, we can map 'good' to 1 and 'bad' to 0
dataset['Quality'] = dataset['Quality'].map({'good': 1, 'bad': 0})
```

```
# Displaying the first few rows of the dataset to confirm the changes
print(dataset.head())
```

```
# Saving the cleaned dataset
dataset.to_csv("/content/apple_quality_cleaned.csv", index=False)
```

# FEATURE SCALING

```
# Importing necessary libraries
from sklearn.preprocessing import StandardScaler
```

```
# Initialize the StandardScaler
scaler = StandardScaler()
```

```
# Selecting only numerical columns for scaling
numerical_cols = ['Size', 'Weight', 'Sweetness', 'Crunchiness', 'Juiciness', 'Ripeness', 'Acidity']
```

```
# Scaling the numerical columns
dataset[numerical_cols] = scaler.fit_transform(dataset[numerical_cols])
```

```
# Displaying the first few rows of the scaled dataset
print(dataset.head())
```

# OUTLIER DETECTION AND HANDLING

```
# Visualizing potential outliers using box plots
import seaborn as sns
import matplotlib.pyplot as plt
```

```
plt.figure(figsize=(10, 6))
sns.boxplot(data=dataset[numerical_cols])
plt.title('Box Plot of Numerical Features')
plt.xlabel('Features')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

```
# Removing outliers (for demonstration purposes)
# You can adjust the threshold and method based on your analysis
outlier_threshold = 3
dataset = dataset[(dataset[numerical_cols] < outlier_threshold).all(axis=1)]
```

```
# Confirming the removal of outliers
print("Number of rows after removing outliers:", len(dataset))
```

# FEATURE ENGINEERING

```
# Creating a new feature by combining existing features
dataset['Combined_Feature'] = dataset['Size'] * dataset['Weight']
```

```
# Displaying the first few rows of the dataset with the new feature
print(dataset.head())
```

```
# Saving the updated dataset with the new feature
dataset.to_csv("/content/apple_quality_cleaned_feature_engineered.csv", index=False)
```
